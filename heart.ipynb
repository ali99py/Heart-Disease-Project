```python
df_cleaned = df.dropna()

print(f"عدد الصفوف قبل الحذف: {len(df)}")
print(f"عدد الصفوف بعد الحذف: {len(df_cleaned)}")

print(df_cleaned.isnull().sum())
```

```python
# حذف عمودي id و dataset
df_cleaned = df_cleaned.drop(columns=['id', 'dataset'])

# عرض الإحصائيات الوصفية للبيانات العددية
print("الإحصائيات الوصفية للبيانات العددية:")
print(df_cleaned.describe())
```

```python
categorical_columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'thal']

plt.figure(figsize=(15,10))

for i, column in enumerate(categorical_columns, 1):
    plt.subplot(3, 3, i)
    df_cleaned[column].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='coolwarm')
    plt.title(f'توزيع {column}')
    plt.ylabel('')

plt.tight_layout()
plt.show()
```

```python
df_encoded = pd.get_dummies(df_cleaned, columns=['sex', 'cp', 'restecg', 'slope', 'thal'], drop_first=True)

from imblearn.over_sampling import SMOTE
X = df_encoded.drop(columns=['num'])
y = df_encoded['num']

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

df_balanced = pd.DataFrame(X_resampled, columns=X.columns)
df_balanced['num'] = y_resampled

print(df_balanced['num'].value_counts())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# رسم توزيع الفئات بعد التوازن
plt.figure(figsize=(8,5))
sns.countplot(x=df_balanced['num'], palette='viridis')

plt.title('Distribution of Target Variable After SMOTE')
plt.xlabel('Heart Disease Class (num)')
plt.ylabel('Count')
plt.show()
```

```python
from imblearn.over_sampling import SMOTE

numerical_columns = df_balanced.select_dtypes(include=['float64', 'int64']).columns

X_numerical = df_balanced[numerical_columns]
y = df_balanced['num']

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_numerical, y)

df_balanced_numerical = pd.DataFrame(X_resampled, columns=numerical_columns)
df_balanced_numerical['num'] = y_resampled

print(df_balanced_numerical['num'].value_counts())
```

```python
from tensorflow.keras.utils import to_categorical

y_one_hot = to_categorical(y, num_classes=5)

```

```python
from sklearn.model_selection import train_test_split

# تقسيم البيانات إلى بيانات تدريب واختبار
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# تحويل y_test إلى one-hot encoding إذا كنت تستخدم التصنيف متعدد الفئات
y_test_one_hot = to_categorical(y_test, num_classes=5)
# تحويل y_train إلى one-hot encoding
y_train_one_hot = to_categorical(y_train, num_classes=5)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam


from tensorflow.keras.layers import Dropout

model = Sequential()
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(5, activation='softmax'))



model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python
history = model.fit(X_train, y_train_one_hot, epochs=50, batch_size=32, validation_data=(X_test, y_test_one_hot))
```

```python
# 11. تقييم النموذج بعد التدريب
score = model.evaluate(X_test, y_test_one_hot, verbose=0)

# عرض الخسارة والدقة
print('خسارة الاختبار:', score[0])
print('دقة الاختبار:', score[1])
```

```python
plt.plot(history.history['accuracy'], label='دقة التدريب')
plt.plot(history.history['val_accuracy'], label='دقة التحقق')
plt.title('دقة النموذج')
plt.xlabel('العصور')
plt.ylabel('الدقة')
plt.legend()
plt.show()
```

```python
# رسم دقة التدريب ودقة التحقق
plt.plot(history.history['accuracy'], label = 'training accuracy')
plt.plot(history.history['val_accuracy'], label='val accuracy')
plt.title('model accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()
plt.show()

# رسم الخسارة أثناء التدريب والتحقق
plt.plot(history.history['loss'], label='training loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.title('model loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix

# افترض أن y_test هي القيم الحقيقية و y_pred هي التنبؤات من النموذج
y_pred = model.predict(X_test)  # التنبؤات من النموذج
y_pred = np.argmax(y_pred, axis=1)  # تحويل التنبؤات إلى فئات إذا كان لديك softmax

# حساب الدقة (Precision)
precision = precision_score(y_test, y_pred, average='weighted')  # average='weighted' لحساب المتوسط الموزون

# حساب الاسترجاع (Recall)
recall = recall_score(y_test, y_pred, average='weighted')

# حساب F1-Score
f1 = f1_score(y_test, y_pred, average='weighted')

# عرض النتائج
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-Score: {f1}")

# مصفوفة التشويش لمزيد من التحليل
conf_matrix = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix:\n{conf_matrix}")
```

```python
from flask import Flask, request, jsonify
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model
from werkzeug.serving import run_simple

# إنشاء التطبيق Flask
app = Flask(__name__)

# تحميل النموذج المحفوظ
model = load_model('heart_model.keras')

# دالة للتنبؤ بالفئة
def predict_class(new_data):
    # تحويل البيانات إلى DataFrame
    new_data_df = pd.DataFrame(new_data)

    # تحويل الفئات إلى one-hot encoding باستخدام pd.get_dummies
    new_data_encoded = pd.get_dummies(new_data_df, columns=['sex', 'cp', 'restecg', 'slope', 'thal'], drop_first=True)

    # استخدام النموذج المحفوظ لتنبؤ المريض الجديد
    predictions = model.predict(new_data_encoded)

    # الفئة المتنبأ بها (أعلى احتمال)
    predicted_class = np.argmax(predictions)
    return predicted_class, predictions

# إنشاء route لواجهة API
@app.route('/predict', methods=['POST'])
def predict():
    # الحصول على البيانات من الـ POST request
    data = request.get_json(force=True)

    # استدعاء الدالة للتنبؤ بالفئة
    predicted_class, predictions = predict_class(data)

    # إرسال النتيجة
    return jsonify({
        'predicted_class': int(predicted_class),
        'predictions': predictions.tolist()
    })

# تشغيل التطبيق باستخدام werkzeug داخل Jupyter
if __name__ == '__main__':
    run_simple('127.0.0.1', 5000, app)

```
